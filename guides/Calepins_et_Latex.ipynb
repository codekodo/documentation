Le format calepin permet l'écriture de formules scientifiques au format Latex. L'ecriure et l'integration de ces formules est tres simple et il est inutile de maitriser l'ensemble de l'environement Latex pour pouvoir utiliser sa syntaxe. Quelques règles suffisent pour pouvoir écrire des formules mathématiques, physiques ou chimiques.
Vous trouverez ci-dessous une synthese des elements pricpaux a maitriser.

# 1. Déclaration Latex

Deux cas. Soit on souhaite ecrire une formule ou une ecriture scientifique simple au sein d'une ligne de texte (presentation en ligne), soit on souhiate presenter l'ecriture scientifique centree dans un nouveau pragaphe (presentation en paragraphe).

## 1.1. Presentation en lgine

Pour une presentation en ligne le code Latex doit etre encadre par des "$".

**Exemple** : `$F(k) = \int_{-\infty}^{\infty} f(x) e^{2\pi i k} dx$` donne $F(k) = \int_{-\infty}^{+\infty} f(x) e^{2\pi i k} dx$
 

## 1.2. Presentation en paragraphe

Pour une presentation en paragraphe, le code Latex doit etre encadré part des `$$` ou par `\begin{xxx} ... \end{xxx}` (avec xxx qui peut être `aligned`, `matrix`, `bmatrix`, `pmatrix`, `equation`, `eqnarray`...).

 
**Exemple 1** :

`$$F(k) = \int_{-\infty}^{+\infty} f(x) e^{2\pi i k} dx$$` donne :
$$F(k) = \int_{-\infty}^{+\infty} f(x) e^{2\pi i k} dx$$

# 2. Intégrer du cde Latex dans un calepin

Le format calepin offre la possiblilite d'intégrer du code Latex dans des cellules "texte". Il est donc possible, grace à Latex,  de presenter des ecritures scientifiques. De la plus simple a la plus complexe.

Ce qui est presente dans ce docuements va se limiter à des ecriture simples et classiques adaptees aux programmes.
 
 Pour integrer du code Latex dans un calepin, il faut donc :
 * créer une cellule "texte"
 * déclarer le code Latex avec `$` ou `$$` ou `\begin{xxx} ... \end{xxx}`
